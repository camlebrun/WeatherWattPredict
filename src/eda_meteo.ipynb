```python
import pandas as pd 
```

```python
df = pd.read_csv('/Users/camille/repo/projet_perso/elec/data/data_2010_2023.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['AAAAMMJJ'], format='%Y%m%d')

# Formater la colonne 'AAAAMMJJ' en "YYYY-MM-DD"
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
df.drop(columns=['AAAAMMJJ'], inplace=True)
```

```python
median_temp_france = df.groupby('Date')['TNTXM'].median().reset_index()
```

```python
import plotly.express as px

fig = px.scatter(median_temp_france, x='Date', y='TNTXM',  title='Diagramme de dispersion de la tempÃ©rature')
fig.show()
```
